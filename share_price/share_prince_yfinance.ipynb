```python
import yfinance as yf
import pandas as pd

# Ticker de Tesla
ticker = 'TSLA'

# Récupération des données de 2022
data = yf.download(ticker, start='2022-01-03', end='2023-01-03')

df = pd.DataFrame(data)
```
